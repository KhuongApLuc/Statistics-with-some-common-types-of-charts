```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('lab2 Topic_Survey_Assignment.csv')
df
```

```python
array = np.array(df)
objects = array[:, 0]

attribute1 = array[:, 1].astype(int)
attribute2 = array[:, 2].astype(int)
attribute3 = array[:, 3].astype(int)

bar_width = 0.25
index = np.arange(len(objects))
fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(index, attribute1, width=bar_width, label='Very interested')
bar2 = ax.bar(index + bar_width, attribute2, width=bar_width, label='Somewhat interested')
bar3 = ax.bar(index + 2 * bar_width, attribute3, width=bar_width, label='Not interested')

ax.set_xlabel('Topics')
ax.set_ylabel('Quantity')
ax.set_title('Bar Chart with Survey about interest in different data science')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(objects, rotation=90)
ax.legend()

plt.xlabel('Topics')
plt.ylabel('Quantity')
plt.title('Bar Chart with Survey about interest in different data science')
plt.xticks(index + bar_width, objects)
plt.xticks(index + bar_width, objects, rotation=90)
plt.legend()

for bar in [bar1, bar2, bar3]:
    for rect in bar:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height, str(height), ha='center', va='bottom')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
index = np.arange(len(objects))
bar_width = 0.35

total = attribute1 + attribute2 + attribute3
percentage1 = (attribute1 / total) * 100
percentage2 = (attribute2 / total) * 100
percentage3 = (attribute3 / total) * 100

bar1 = plt.bar(index, percentage1, label='Very interested', width=bar_width)
bar2 = plt.bar(index, percentage2, bottom=percentage1, label='Somewhat interested', width=bar_width)
bar3 = plt.bar(index, percentage3, bottom=percentage1+percentage2, label='Not interested', width=bar_width)

plt.xlabel('Topics')
plt.ylabel('Quantity')
plt.title('Stacked Bar Chart with Survey about interest in different data science topics')
plt.xticks(index, objects, rotation=90)

plt.legend()
plt.show()
```

```python
attributes = array[:, 1:].astype(int)

fig, axs = plt.subplots(3,2, figsize=(10, 10))

for i, (obj, attrs) in enumerate(zip(objects, attributes)):
    row, col = divmod(i, 2)
    axs[row, col].pie(attrs, labels=['Very interested', 'Somewhat interested', 'Not interested'], autopct='%1.1f%%', startangle=90)
    axs[row, col].set_title(obj)

fig.suptitle('Pie Charts for each topics', fontsize=16)
fig.text(0.5, 0.05, 'Interest Levels', ha='center', fontsize=14)
plt.show()
```
